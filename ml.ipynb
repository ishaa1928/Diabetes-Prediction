```python
y=df["Outcome"]
X=df.drop("Outcome",axis=1,)
```

```python
from sklearn.model_selection import train_test_split
```

```python
pip install sklearn
```

```python
pip install scikit-learn
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression()
dr = DecisionTreeClassifier()
```

```python
lr.fit(X_train,y_train)
dr.fit(X_train,y_train)
```

```python
print("logistic",lr.score(X_train,y_train),lr.score(X_test,y_test))
print("decision tree",dr.score(X_train,y_train),dr.score(X_test,y_test))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(dr,X_test,y_test,cv=5)
```

```python
cross_val_score(lr,X_train,y_train,cv=5)
```

```python
cross_val_score(lr,X_test,y_test,cv=5)
```

```python
param_grid = {
    'C' : [0.01,0.1,1,10,100,1000],           #regularisation parameter
    'penalty':['l2','l1','elastinet'],   #penalty types
    'solver':['liblinear']  # Optimization algorithm
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Setup GridSearch with cross-val 
grid_search = GridSearchCV(estimator = lr,param_grid=param_grid,cv=5,scoring='accuracy')

#fit the model with the training data
grid_search.fit(X_train,y_train)

#best hyperparametry found
best_params = grid_search.best_params_
print("Best Parameters:",best_params)
```
